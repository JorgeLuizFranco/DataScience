````python

import geopandas as gpd
from shapely.geometry import Point, Polygon
import pandas as pd
from fiona.drvsupport import supported_drivers
supported_drivers['KML'] = 'rw'
my_map = gpd.read_file('/home/Data_Ovinos/sandra.kml', driver='KML')
#print(my_map)
macho=pd.read_csv('/home/Data_Ovinos/OvisMacho1.csv')
femea1=pd.read_csv('/home/Data_Ovinos/OvisFemea1.csv')
femea2=pd.read_csv('/home/Data_Ovinos/OvisFemea2.csv')
macho
````

Transformando em dados geométricos (geopandas)

````python
ovinos=[femea1,femea2,macho]

ovinos_geo_locations=[]
ovinos_geometric_points=[]
for ovino in ovinos:
    geometric_points = []
    for xy in zip(ovino[' Longitude'], ovino[' Latitude']):
        geometric_points.append(Point(xy))
    ovinos_geometric_points.append(geometric_points)

geo_locations = gpd.GeoDataFrame(macho,crs = {'init': 'epsg:4326'},geometry = geometric_points)
for i in range(len(ovinos)):
  ovinos_geo_locations.append(gpd.GeoDataFrame(ovinos[i],crs = {'init': 'epsg:4326'},geometry = ovinos_geometric_points[i]))
#print(ovinos_geo_locations)
````

````python
print(geo_locations.head(5))
import folium
import json
geo_json_map=json.load(open('/home/Data_Ovinos/sandra.geojson'))
map_markers=folium.Map(zoom_start=100,location=[-18.988649, -56.618572],min_zoom=70)

folium.Choropleth(
    geo_data = geo_json_map,
    fill_color = "steelblue",
    fill_opacity = 0.4,
    line_color = "steelblue",
    line_opacity = 0.9
).add_to(map_markers)

points = geo_locations[[' Latitude', ' Longitude']].values.tolist()
#print(points)
for point in range(len(points)//100):
    #print(points[point])
    folium.Marker(location=points[point],popup=str(point+1),icon=folium.Icon(color='blue',icon=point)).add_to(map_markers)

map_markers.save('/home/mapa.html')
````

# Pastagens mais frequentes


````python
from scipy.spatial import distance
sorted_tuples=[]
for i in range(len(name_files)):
  frequences={}
  #print(name_files[i])
  #print(coords_ovelhas[i])
  for index,pastagem in df_pastagens.iterrows():
    name_past= pastagem['Sp chave 1'] + ' --- ' +str(pastagem['Sp chave 2']);
    frequences[name_past]=0
  for coord in coords_ovelhas[i]:
    menor= ['',1000000000000000000000]
    for index,pastagem in df_pastagens.iterrows():
      name_past= pastagem['Sp chave 1'] + ' --- ' +str(pastagem['Sp chave 2']);
      p1= [pastagem['X'],pastagem['Y']]
      p2=coord
      dist = distance.euclidean(p1,p2)
      if(dist< menor[1]):
        menor[1]=dist
        menor[0]=name_past

    frequences[menor[0]]+=1
  sorted_tuples.append(sorted(frequences.items(), key=lambda item: item[1])) 
  #print(sorted_tuples[i])  

  
      
    
  
````

Escrevendo arquivos

````python
freq_pasto_tuples= []
for t in sorted_tuples:
  #print(t)
  freq_pasto_tuples.append(tuple(t))
#print(freq_pasto_tuples)
for i in range(len(name_files)):
  with open('/home/Data_Ovinos/frequencias_pasto_ovelha_'+name_files[i]+'.txt', 'w') as fp:
      fp.write(name_files[i]+':\n\n')
      print(name_files[i]+':\n\n')
      for pastag in freq_pasto_tuples[i]:
        print(str(pastag)+'\n')
        fp.write(str(pastag)+'\n')
      
````
